Visualisasi Scatter Plot Perbandingan Kuantitatif

Pada tugas kali ini kita akan mengamati nilai DewPointAvg (F) dengan mengamati nilai HumidityAvg (%), TempAvg (F), dan WindAvg (MPG)

Perhatikan bahwa data kita tidaklah siap untuk di analisis, salah satunya tipe data dari DewPointAvg (F), HumidityAvg (%), dan WindAvg (MPG) adalah object, padahalnya data nya ber isi numeric. maka :

Ubahlah tipe data tersebut menjadi tipe data float

Step2 :

Kalian tidak akan dengan mudah mengubah tipe data tersebut karena column tersebut mempunyai nilai '-' yang dimana tidak bisa di ubah ke bentuk float, maka replace lah terlebih dahulu data yang bernilai '-' dengan nilai NaN, gunakan method .replace(). baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html
Isi nilai nan dengan nilai sebelumnya di row tersebut. gunakan method .fillna() dengan argument method bernilai 'ffill', baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html
Sekarang ubah tipe datanya dengan float, gunakan method .astype(), baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html Setelah ini sebagian data siap untuk di jadikan bahan analisis. maka :
Buahlah visualisasi perbandingan kuantitatif scatter plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('austin_weather.csv')
df.head()
```

```python
df.info()
```

```python
df['VisibilityLowMiles'].unique()
```

```python
df['VisibilityLowMiles'].replace ('-', np.nan, inplace = True)
df['VisibilityLowMiles'].fillna(method ='ffill', inplace = True)
df['VisibilityLowMiles']= df['VisibilityLowMiles'].astype ('float')
df['VisibilityLowMiles'].unique()
df.info()
```

```python
def preprocess_data(series):
    series.replace('-', np.nan, inplace = True)
    series.fillna(method='ffill', inplace = True)
    return series.astype ('float')

df['DewPointAvgF'] = preprocess_data(df['DewPointAvgF'])
df['HumidityAvgPercent'] = preprocess_data(df['HumidityAvgPercent'])
df['WindAvgMPH'] = preprocess_data(df['WindAvgMPH'])

fig,ax = plt.subplots(figsize= (14,8))
mapbar = ax.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], cmap='coolwarm', 
                    c = df['DewPointAvgF'], s = df['WindAvgMPH']*5)

fig.colorbar(mapbar)
ax.set_title('Asustin Weather')
ax.set_xlabel('HumidityAvg(%)')
ax.set_ylabel('DewPointAvg(F)')
plt.show()
```
